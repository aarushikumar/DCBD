```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
```

```python
def fibonacci(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fibonacci(n-1)+fibonacci(n-2)
```

```python
def sumofn(n):
    if n==1:
        return 1
    return n+sumofn(n-1)
```

```python
sumofn(10)
```

```python
def printN(n):
    if n==0:
        return 
    print(n)
    
    printN(n-1)
```

```python
def reverse(s):
    if len(s)==0:
        return s
    return reverse(s[1:])+s[0]
```

```python
reverse("hello")
```

```python
def printarray(arr):
    if len(arr)==0:
        return None
    print(arr[-1])
    printarray(arr[:len(arr)-1])
```

```python
printarray([1,2,3,4,5])
```
